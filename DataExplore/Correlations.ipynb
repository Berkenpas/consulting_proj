```python
import os
import sys
from os.path import join

import pandas as pd
```

```python
data_dir = "../private/coins"

def load_data(data_dir): 
    coin_data = {}
    for file_name in os.listdir(data_dir): 
        file_data = pd.read_csv(join(data_dir, file_name))
        coin_data[file_name.split(".")[0]] = file_data
    
    return coin_data

coin_data = load_data(data_dir)
del coin_data['cfx']
del coin_data['vgx']
del coin_data['prom']



print(coin_data.keys())
        
```

```python
display(coin_data['btc'])
```

```python
#pearson correlation
from scipy.stats import pearsonr


coin_corr_pearson = {}

for coin1 in coin_data: 
    for coin2 in coin_data: 
        if coin1 in coin2: 
            continue 
        
        coin1_data = coin_data[coin1]['price_high'].values
        coin2_data = coin_data[coin2]['price_high'].values
        
        length = min(len(coin1_data), len(coin2_data))
        
#         print(len(coin1_data[:length]))
#         print(len(coin2_data[:length]))

        corr = pearsonr(coin1_data[:length], coin2_data[:length])
    
        coin_corr_pearson[":".join([coin1, coin2])] = corr
        
for pair, corr in coin_corr_pearson.items(): 
    print(f"{pair}: corr: {corr}")
    print('\n')
    print("-"*75)
```

```python
#kendall_tau correlation
from scipy.stats import kendalltau


coin_corr_kendall = {}

for coin1 in coin_data: 
    for coin2 in coin_data: 
        if coin1 in coin2: 
            continue 
        
        coin1_data = coin_data[coin1]['price_high'].values
        coin2_data = coin_data[coin2]['price_high'].values
        
        length = min(len(coin1_data), len(coin2_data))
        
#         print(len(coin1_data[:length]))
#         print(len(coin2_data[:length]))

        corr = kendalltau(coin1_data[:length], coin2_data[:length])
    
        coin_corr_kendall[":".join([coin1, coin2])] = corr
        
for pair, corr in coin_corr_kendall.items(): 
    print(f"{pair}\ncorr: {corr}")
    print('\n')
    print("-"*75)
```

```python
#kendall_tau correlation
from scipy.stats import spearmanr


coin_corr_spearman = {}

for coin1 in coin_data: 
    for coin2 in coin_data: 
        if coin1 in coin2: 
            continue 
        
        coin1_data = coin_data[coin1]['price_high'].values
        coin2_data = coin_data[coin2]['price_high'].values
        
        length = min(len(coin1_data), len(coin2_data))
        
#         print(len(coin1_data[:length]))
#         print(len(coin2_data[:length]))

        corr = spearmanr(coin1_data[:length], coin2_data[:length])
    
        coin_corr_spearman[":".join([coin1, coin2])] = corr
        
for pair, corr in coin_corr_spearman.items(): 
    print(f"{pair}\ncorr: {corr}")
    print('\n')
    print("-"*75)
```

```python
#Finding the highest correlations 
baby_coins = ["DENT", "REN", "VGX", "AR", "1INCH", "XVG", "CRV", "PROM", "CFX", "LSX"]
established_coins = ["BTC", "ETH", "BNB", "ADA", "DOT", "UNI", "LTC", "LINK", "VET", "XLM"]
#1. Pearson correlation 
highest_pearson_correlation = {}
for k, v in coin_corr_pearson.items(): 
    if k.split(":")[0].upper() in baby_coins and not k.split(":")[1].upper() in baby_coins: 
        highest_pearson_correlation[k] = v
        
to_delete = []
highest_pearson_correlation = {k: v for k, v in sorted(highest_pearson_correlation.items(), key=lambda item: abs(item[1][0]), reverse=True)}
for k, v in highest_pearson_correlation.items(): 
    if v[1] > 0.05: 
        to_delete.append(k)
        
for k in to_delete: 
    del highest_pearson_correlation[k]

    
#Uncomment this line to see the whole correlations
# print(highest_pearson_correlation)


pearson_high_counts = {}
for k, v in highest_pearson_correlation.items(): 
    coin = k.split(':')[0]
    if coin in pearson_high_counts: 
        pearson_high_counts[coin] += 1
    else: 
        pearson_high_counts[coin] = 1
        
        
print(pearson_high_counts)
```
